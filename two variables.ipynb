# Two Variable Plots
adapted from https://matplotlib.org/tutorials/introductory/pyplot.html

```python
import matplotlib.pyplot as plt
import numpy as np
```

y = 2x

```python
x = np.linspace(0.0, 100.0, 11)
y = x * x
y2 = x * x * x
y3 = x * x * x * x
# x squared in comparison to x cubed
plt.plot(x, y, 'r')
plt.plot(x, y2, 'b')
plt.plot(x, y3, 'g')
```
